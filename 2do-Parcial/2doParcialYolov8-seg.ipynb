```python
import glob
from IPython.display import Image, display

# ObtÃ©n la lista de rutas de las imÃ¡genes
imagen_rutas = glob.glob('/content/drive/MyDrive/IA2-2-2023/2do-Parcial/dataset/runs/segment/predict/*.jpg')

# Itera sobre cada 30 imÃ¡genes y muestra las primeras 6
for i in range(0, len(imagen_rutas), 30):
    for j in range(6):
        idx = i + j
        if idx < len(imagen_rutas):
            image_path = imagen_rutas[idx]
            display(Image(filename=image_path, width=600))
            print("\n")
```
